```python
# plot distrubtion of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=10,kde=True)
plt.title("distrubtion of generated power(KW)")
plt.xlabel("generated power(KW)")
plt.ylabel('frequence')
plt.show()
```

```python
# bivariate analysis
# scatter plot with the target feature
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
    plt.subplot(7, 3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Generated power (KW)')
    plt.xlabel(column)
    plt.ylabel('Generated power (KW)')
plt.tight_layout()
plt.show()
```
